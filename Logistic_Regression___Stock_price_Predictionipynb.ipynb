Logistic Regression predicts a binary output , For Eg - Is the mail spam or not not , while in our case 
Will the stock Market go up or down on the next day ?

```python
import pandas as pd
import statsmodels.api as sm
```

```python
data = pd.read_csv('/content/RELIANCE.NS (1).csv')
```

```python
data_reset=data.set_index('Date')
```

```python
data_reset
```

```python
df = data_reset['Adj Close'].pct_change()*100
# for getting values in %
```

```python
df
```

```python
df = df.rename("Today")
```

```python
df
```

```python
df = df.reset_index()
```

```python
df
```

```python
for i in range(1,6):
  df['Lag '+str(i)] = df['Today'].shift(i)
```

```python
df
```

```python
df['Volume']= data.Volume.shift(1).values/1_000_000
```

```python
df = df.dropna()
```

```python
df
```

```python
#1  -- up and 0 -- down 
df['Direction'] = [1 if i > 0 else 0 for i in df['Today']]
```

```python
df
```

```python
 df = sm.add_constant(df)
```

```python
df
```

```python
# here by using the lags and volume , we want to predict the direction i.e 1 or 0 
```

```python
# the independent varaible are lags & volume
# the dependent varaible is direction
```

```python
df
```

```python
X= df[['const','Lag 1','Lag 2','Lag 3','Lag 4','Lag 5','Volume']]
```

```python
y= df.Direction
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
# spliting the train and test dataset
x_value= df.iloc[0:928,2:8]
y_value=df.iloc[0:928,9:10]
y_value
x_test=df.iloc[928:,2:8]
y_test=df.iloc[928:,9:10]
```

```python
#model=sm.Logit(x_train,y_train)
#data = x_train[pd.notnull(x_train)]
#data1=y_train[pd.notnull(y_train)]
```

```python
from sklearn.linear_model import LogisticRegression

# instantiate the model (using the default parameters)
logreg = LogisticRegression()

# fit the model with data
logreg.fit(x_value,y_value)

#
y_pred=logreg.predict(x_test)
```

```python
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python

#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```
